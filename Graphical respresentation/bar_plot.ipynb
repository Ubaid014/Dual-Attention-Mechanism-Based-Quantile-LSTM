# Hybrid vs conventionl MAE PLOT

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
d = {'LSTM': [46.01,20.09], 'QLSTM(0.5)': [47.37,23.73], 'QLSTM(0.6)': [53.80,22.68], 'QLSTM(0.9)': [94.26,44.97],'DA-QLSTM': [41.61,19.17]}
data=pd.DataFrame(data=d, index=["IESCO","Panama"])
print(data)
x1=data['LSTM']
x2=data['QLSTM(0.5)']
x3=data['QLSTM(0.6)']
x4=data['QLSTM(0.9)']
x5=data['DA-QLSTM']

#x4=data['LSTM']
y=["IESCO","Panama"]
sns.set_style("white")
y_axis=np.arange(len(y))
hatches = ['*', '.', 'x', '/','o']

plt.rcParams['font.size'] = '16'
plt.rcParams["font.family"] = "Times New Roman"
plt.legend(prop={"size":14})
plt.rcParams["figure.figsize"] = (8,6)
plt.ylabel("MAE (kW)",fontweight="bold")
plt.bar(y_axis, x5,0.1, label= 'DA-QLSTM' ,edgecolor='black', color=['mistyrose'], hatch=hatches[0])
plt.bar(y_axis+0.1, x2,0.1, label= 'QLSTM(0.5)' ,edgecolor='black', color=['royalblue'], hatch=hatches[1])
plt.bar(y_axis+0.2, x3,0.1, label= 'QLSTM(0.6)',edgecolor='black',color=['sandybrown'], hatch=hatches[2])
plt.bar(y_axis+0.3, x1,0.1, label= 'LSTM',edgecolor='black',color=['darkseagreen'], hatch=hatches[3])
plt.bar(y_axis+0.4, x4,0.1, label= 'QLSTM(0.9)',edgecolor='black',color=['silver'], hatch=hatches[4])

#plt.bar(y_axis+0.6, x4,0.2, label= 'LSTM',edgecolor='black', color=['darkseagreen'],hatch=hatches[3] )
plt.xticks(y_axis, y)
plt.grid(color = 'black', linestyle = '--', linewidth = 0.2)
#plt.xlabel("Years")
plt.legend()
plt.savefig('Hybrid_vs_conventional_MAE.pdf',dpi=300) 
plt.savefig('Hybrid_vs_conventional_MAPE.eps',dpi=300) 
plt.savefig('Hybrid_vs_conventional_MAPE.png',dpi=300) 

```

# MAPE with SOTA MODELS 

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
d = {'GRU': [4.32, 1.70], 'RNN': [4.71,1.81], 'RNN-GRU': [4.29,1.73], 'RNN-LSTM': [4.61,1.72],'DA-QLSTM': [4.06,1.66]}
data=pd.DataFrame(data=d, index=["IESCO","Panama"])
print(data)
x1=data['GRU']
x2=data['RNN']
x3=data['RNN-GRU']
x4=data['RNN-LSTM']
x5=data['DA-QLSTM']

#x4=data['LSTM']
y=["IESCO","Panama"]
sns.set_style("white")
y_axis=np.arange(len(y))
hatches = ['*', '.', 'x', '/','o']

plt.rcParams['font.size'] = '16'
plt.rcParams["font.family"] = "Times New Roman"
plt.legend(prop={"size":14})
plt.rcParams["figure.figsize"] = (8,6)
plt.ylabel("MAPE (%)",fontweight="bold")
plt.bar(y_axis, x5,0.1, label= 'DA-QLSTM' ,edgecolor='black', color=['mistyrose'], hatch=hatches[0])
plt.bar(y_axis+0.1, x2,0.1, label= 'GRU' ,edgecolor='black', color=['royalblue'], hatch=hatches[1])
plt.bar(y_axis+0.2, x3,0.1, label= 'RNN',edgecolor='black',color=['sandybrown'], hatch=hatches[2])
plt.bar(y_axis+0.3, x1,0.1, label= 'RNN-GRU',edgecolor='black',color=['darkseagreen'], hatch=hatches[3])
plt.bar(y_axis+0.4, x4,0.1, label= 'RNN-LSTM',edgecolor='black',color=['silver'], hatch=hatches[4])

#plt.bar(y_axis+0.6, x4,0.2, label= 'LSTM',edgecolor='black', color=['darkseagreen'],hatch=hatches[3] )
plt.xticks(y_axis, y)
plt.grid(color = 'black', linestyle = '--', linewidth = 0.2)
#plt.xlabel("Years")
plt.legend()
plt.savefig('MAPE_VS_SOTA.pdf',dpi=300) 
plt.savefig('MAPE_VS_SOTA.eps',dpi=300) 
plt.savefig('MAPE_VS_SOTA.png',dpi=300) 

```
