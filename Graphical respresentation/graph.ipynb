# PANAMA GRAPHICALL REPRESENTATION

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

# Load the data
df= pd.read_csv("PREDICTION PANAMA.csv")
df.head()
```

```python

df= df.head(100)
# Set up the main plot
plt.rcParams['font.size'] = '16'
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["figure.figsize"] = (8, 6)
plt.plot(df['ACTUAL'], label= 'Measured', marker = '+', linewidth = 2)
#plt.plot(df['LSTM'],label='LSTM')
#plt.plot(df['GRU'], label= 'GRU')
#plt.plot(df['RNN-LSTM'], label= 'RNN-LSTM')
#plt.plot(df['RNN-GRU'],label='RNN-GRU')
#plt.plot(df['Q-LSTM'],label='Q-LSTM' )
plt.plot(df['PROPOSED APPROACH'],label='DA-QLSTM', marker = '*',linewidth = 2)
plt.grid(color = 'black', linestyle = '--', linewidth = 0.2)
plt.xlabel('Time (h)',fontweight='bold')
plt.ylabel("LOAD (kW)", fontweight='bold')
plt.legend(prop={"size":14} )
plt.title('Panama Dataset', fontweight='bold')


plt.savefig('Graphicall_OUTPUT_PANAMA.pdf',bbox_inches='tight',dpi=300)
plt.savefig('Graphicall_OUTPUT_PANAMA.eps',bbox_inches='tight',dpi=300)
plt.savefig('Graphicall_OUTPUT_PANAMA.png',bbox_inches='tight',dpi=300)
# Show the plot
plt.show()
```
